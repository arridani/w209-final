```python
starting_pitchers = []
for g in tqdm(df.game_pk.unique()):#g in tqdm(df.game_date.unique()):
    sub_df = df[(df.game_pk == g) & (df.inning==1)]
#     for ht in sub1.home_team.unique():
#         for at in sub1.away_team.unique():
#             sub_df = df[(df.game_date == g) &
#                        (df.home_team == ht) &
#                        (df.away_team == at) &
#                        (df.inning == 1)]
    if sub_df.empty: continue
    home_pitcher = sub_df[sub_df.inning_topbot == 'Top'].player_name.unique()
    away_pitcher = sub_df[sub_df.inning_topbot == 'Bot'].player_name.unique()
    if len(home_pitcher) == len(away_pitcher):
        for i in range(len(home_pitcher)):
            home_pitcher_pitches = df[(df.game_date == g) &
                       (df.home_team == ht) &
                       (df.away_team == at) &
                       (df.player_name == home_pitcher[i])].shape[0]
            away_pitcher_pitches = df[(df.game_date == g) &
                       (df.home_team == ht) &
                       (df.away_team == at) &
                       (df.player_name == away_pitcher[i])].shape[0]
            obj = {'game_date': g,
                  'home_team': ht,
                  'away_team': at,
                  'home_pitcher': home_pitcher[i],
                  'away_pitcher': away_pitcher[i],
                  'home_pitcher_pitches': home_pitcher_pitches,
                  'away_pitcher_pitches': away_pitcher_pitches}
            starting_pitchers.append(obj)
    elif len(home_pitcher) > len(away_pitcher):
        for i in range(len(away_pitcher)):
            home_pitcher_pitches = df[(df.game_date == g) &
                       (df.home_team == ht) &
                       (df.away_team == at) &
                       (df.player_name == home_pitcher[i])].shape[0]
            away_pitcher_pitches = df[(df.game_date == g) &
                       (df.home_team == ht) &
                       (df.away_team == at) &
                       (df.player_name == away_pitcher[i])].shape[0]
            obj = {'game_date': g,
                  'home_team': ht,
                  'away_team': at,
                  'home_pitcher': home_pitcher[i],
                  'away_pitcher': away_pitcher[i],
                  'home_pitcher_pitches': home_pitcher_pitches,
                  'away_pitcher_pitches': away_pitcher_pitches}
            starting_pitchers.append(obj)
    elif len(home_pitcher) < len(away_pitcher):
        for i in range(len(home_pitcher)):
            home_pitcher_pitches = df[(df.game_date == g) &
                       (df.home_team == ht) &
                       (df.away_team == at) &
                       (df.player_name == home_pitcher[i])].shape[0]
            away_pitcher_pitches = df[(df.game_date == g) &
                       (df.home_team == ht) &
                       (df.away_team == at) &
                       (df.player_name == away_pitcher[i])].shape[0]
            obj = {'game_date': g,
                  'home_team': ht,
                  'away_team': at,
                  'home_pitcher': home_pitcher[i],
                  'away_pitcher': away_pitcher[i],
                  'home_pitcher_pitches': home_pitcher_pitches,
                  'away_pitcher_pitches': away_pitcher_pitches}
            starting_pitchers.append(obj)
    else:
        raise ValueError

# starting_pitchers[['game_date', 'release_speed', 'player_name', 'pitcher', 
#        'p_throws', 'home_team', 'away_team',
#        'outs_when_up', 'inning', 'inning_topbot', 'pitch_number', 'pitch_name',
#        'home_score', 'away_score', 'bat_score', 'fld_score', 'post_away_score',
#        'post_home_score', 'post_bat_score', 'post_fld_score',
#        'if_fielding_alignment', 'of_fielding_alignment', 'spin_axis',
#        'delta_home_win_exp', 'delta_run_exp']]
```
