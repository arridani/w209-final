```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
plt.style.use('ggplot')
```

```python
df = pd.read_csv('../data/statcast_04_04_2022_10_03_2022.csv')
```

```python
# filter columns
columns = ['game_date', 'player_name', 'batter', 'pitcher', 'events',
       'description', 
       'stand', 'p_throws', 'home_team', 'away_team', 'release_spin_rate',
       'outs_when_up', 'inning', 'inning_topbot', 'at_bat_number', 'pitch_number', 'pitch_name',
       'home_score', 'away_score']
filtered = df[columns]
filtered['game_date'] = pd.to_datetime(filtered.game_date)
print(df.columns)
filtered
```

```python
filtered.to_csv('../data/spin_rate.csv', index = 0)
```

```python
grouped_df = filtered.groupby('game_date').mean()['release_spin_rate'].reset_index()
sns.lineplot(grouped_df.game_date, grouped_df.release_spin_rate)
plt.show()
```
